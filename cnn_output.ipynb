```python
from keras.applications.resnet50 import ResNet50
from keras.preprocessing import image
from pathlib import Path
import keras
import numpy as np

folder_path=Path("Flickr8k/captions/")
train_path=folder_path/"Flickr_8k.trainImages.txt"
image_dir=Path("Flickr8k/Flicker8k_Dataset/")
```

```python
model=ResNet50(include_top=False,weights="imagenet",input_shape=(224,224,3),pooling="avg")
model.summary()
```

```python
def preprocess(path):
    imag=image.load_img(path,target_size=(224,224,3))
    imag=image.img_to_array(imag)
    imag=np.expand_dims(imag,axis=0)
    return imag
```

```python
with open(train_path) as file:
    train_set=file.read().split("\n")
    
train_pred={}    
checkpt=0    
for idx in train_set:
    if idx=="":
        continue
    checkpt+=1    
    if checkpt%1000==0:
        print(checkpt)
    path=image_dir/idx
    img=preprocess(path)
    prediction=model.predict(img)
    train_pred[idx]=prediction
```
